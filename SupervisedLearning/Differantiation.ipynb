```
import numpy as np
def derivative(f,x):
  delta_x=1e-4
  return (f(x+delta_x)-f(x-delta_x))/(2*delta_x)
def myFunc1(x):
  return x**2

result=derivative(myFunc1,3)
print("result == ",result)
def myFunc2(x):
  return 3*x*np.exp(x)
result2=derivative(myFunc2,2)
print("result == ",result2)
# numerical derivative 편미분
def myFunc3(x,y):
  return 2*x+3*x*y+y**3

def numerical_derivative(f,x):# x와 y에 대한 편미분은 (y고정 x변화 수치, x고정 y변화 수치)
#여기서는 x로 변수를 모두 포함한다
  delta_x=1e-4
  grad=np.zeros_like(x)# x의 형태와 동일하게 gradient (결과값)저장
  it=np.nditer(x,flags=['multi_index'],op_flags=['readwrite'])
  while not it.finished:
    idx=it.multi_index
    tmp_val=x[idx] # iter번째 요소를 잠시 저장 
    x[idx]=float(tmp_val)+delta_x
    fx1=f(x)#요소 delta변화 후 저장
    x[idx]=float(tmp_val)-delta_x
    fx2=f(x)#요소 delta변화 후 저장
    grad[idx]=(fx1-fx2)/(2*delta_x)
    x[idx]=tmp_val
    it.iternext()
  return grad



```
